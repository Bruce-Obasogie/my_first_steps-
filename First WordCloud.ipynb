```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import pandas as pd

#content-related part
texts= open(r"C:\Users\USER\Documents\batman_script.txt").read() 
stopwords = set(STOPWORDS)


#appearance-related part
wc= WordCloud(background_color= 'red', max_font_size = 50, stopwords= stopwords)
wc.generate(texts)

#plotting section
plt.imshow(wc, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```
